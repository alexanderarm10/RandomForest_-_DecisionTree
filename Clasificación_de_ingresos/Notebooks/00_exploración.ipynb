# Carga de datos

```python
# librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# carga de datos
columnas = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
            'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
            'hours-per-week', 'native-country', 'income' ]
df_ingresos = pd.read_csv('../Datos/Crudos/census+income/adult.data', header=None, names=columnas)

df_ingresos.head()
```

```python
# Resúmen de los datos
print(df_ingresos.info())
print(f'\n{df_ingresos.dtypes.value_counts()}')
```

```python
# Cantidad de nulos
df_ingresos.isnull().sum()
```

```python
# Estadísticos de la data numércia
df_ingresos.describe()
```

```python
# Balanceo de variable objetivo
print(df_ingresos['income'].value_counts())
sns.countplot(data=df_ingresos, x='income', hue='income', palette='Blues', edgecolor='black')
plt.show();
```

```python
# Guardar data
df_ingresos.to_csv('../Datos/Procesados/income.csv')
```
